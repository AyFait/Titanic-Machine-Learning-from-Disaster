```python
#importing req modules
from xgboost import XGBClassifier
```

```python
#loading data
trainFile = '/kaggle/input/house-prices-art-cleaned/Titanic-MchLrnfrmDis-TrainCln.csv'
testFile = '/kaggle/input/house-prices-art-cleaned/Titanic-MchLrnfrmDis-TestCln.csv'
submissionFile = '/kaggle/working/submission.csv'
train = pd.read_csv(trainFile)
test = pd.read_csv(testFile)
#train.head(5)
print(train[:6])
#test.head(5)
print(test[:6])
```

```python
X_train = train.drop('Survived', axis=1)
#X_train.head(5)
print(X_train[:6])
y_train = train['Survived']
#y_train.head(5)
print(y_train[:6])
X_test = test
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)#
yPred = model.predict(X_test)
print(X_test[:6])
print(yPred[:6])
```

```python
sub = {'PassengerId': X_test['PassengerId'],
      'Survived': yPred}
```

```python
#pID = X_test['PassengerId']
submission = pd.DataFrame(sub)
print(submission[:6])
submission.to_csv(submissionFile, index = False)
print('Submission downloased successfully!')
```
